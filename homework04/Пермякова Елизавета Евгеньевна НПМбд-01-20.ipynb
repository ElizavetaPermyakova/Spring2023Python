11.  Отнимите от каждого элемента матрицы $A$ среднее по строке и поделите на стандартное отклонение строки.

```python
M = np.mean(A, axis = 1)
o = np.std(A, axis = 1)
for i in range(rows):
    for j in range(columns):
        A[i][j] -= M[i]
        A[i][j] /= o[i]
print(A)
```

```python
np.random.seed(1032201728)
np.random.randint(9, size=1)
```

```python
def f0(x):
    return (x**2 + 3) / (x - 2)

def f9(x):
    return 1 / np.sqrt(2+3*x-2*x**2, where = x!=2)

def solution(func, a, b, dx, method):
    n = int((b - a) / dx)
    A = np.linspace(a, b, n)
    if method == 'rect':
        return dx * np.sum(func(A))
    elif method == 'trapeze':
        return dx / 2 * (func(A[0]) + 2 * np.sum(func(A[1:-1])) + func(A[-1]))
    elif method == 'simpson':
        return dx / 3 * (func(A[0]) + 2 * np.sum(func(A[2:-1:2])) + 4 * np.sum(func(A[1:-2:2])) + func(A[-1]))
    else:
        return 'wrong method'
    
AnSol0 = (11/2) + 7 * np.log(2)
AnSol9 = np.pi * np.sqrt(2) / 4

print(f"Аналитические решения: №8: {AnSol0}, №9: {AnSol9}")
print(f"Метод прямоугольников: №8: {solution(f0, 3, 4, 0.0001, 'rect')}, Для №9: {solution(f9, 3/4, 2, 0.0001, 'rect')}")
print(f"Метоц трапеций: №8: {solution(f0, 3, 4, 0.0001, 'trapeze')}, Для №9: {solution(f9, 3/4, 2, 0.0001, 'trapeze')}")
print(f"Метод Симпсона: №8: {solution(f0, 3, 4, 0.0001, 'simpson')}, Для №9: {solution(f9, 3/4, 2, 0.0001, 'simpson')}")
```
